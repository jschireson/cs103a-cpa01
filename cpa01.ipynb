Some analysis of the chosen passwords:
    - The biggest differentiator seemed to be the numer of digits/characters in the password. In the above seaborn plots and the immediate above chart, we can see that very high digit passwords didn't contain any letters. Passwords with more digits also decreased drastically in length at 4 digits. Passwords with 3 digits are likely longer on average because many of these are a combination of letters with 123 added (4.3 non digit characters on average). Passwords with 4 digits are more likely to be just numbers, as the average length was 4.8 likely many of the passwords were just a 4 digit combination. Also interesting is the near zero number of characters at 8 digits, shown above and in some histograms. Certain websites require 8 characters as a minimum threshold for password security, but may allow for passwords of exclusively digits. Almost no passwords (likely just 1 or 2) contained 8 letters and any characters, while similarly long 7 and 9 digit passwords contained ~50x-150x more characters relative to 8 letter passwords.
